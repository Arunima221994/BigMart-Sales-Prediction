```python
## outlier checking
def find_outlier (col_name):
    Q1=df[col_name].quantile(0.25)
    Q3=df[col_name].quantile(0.75) 
    IQR=Q3-Q1
    
outliers=df[(df[col_name]<Q1 -1.5*IQR) | (df[col_name]> Q3 -1.5*IQR)]

print ("Total count of outliers are", outliers.shape[0])
```

```python
## selecting target variables and features

y=df['Item_Outlet_Sales']
x=df.iloc[0:8523,1:12].values

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train)
print(y_test)
```

```python
## Linear Regression

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import VotingRegressor
lin_reg=LinearRegression()

## fitting the model
lin_reg.fit(x_train,y_train)

## predicted the model
y_pred=lin_reg.predict(x_test)
y_pred

```

```python
lin_reg.score(x_test,y_test)
```

```python
lin_reg.score(x_train,y_train)
```

```python

r_squared = r2_score(y_test, y_pred)
MAE = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)  # Calculate RMSE using MSE

# Print the metrics
print(f'r_squared: {r_squared}')
print(f'MAE: {MAE}')
print(f'MSE: {MSE}')
print(f'RMSE: {RMSE}')
```
